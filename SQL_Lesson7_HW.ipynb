## Вариант 1. C использованием возможностей pandas (df.to_sql)

```python
#task1  (lesson7)
#sqlite3: Сделать тестовый проект с БД (sqlite3, project name: task1_7). В таблицу table1 записать 1000 строк
#с случайными значениями (3 колонки, тип int) от 0 до 1000.
#Далее построить гистаграмму распределения этих трех колонок

import sqlite3
import pandas as pd
import plotly.express as px
from random import randint

conn = sqlite3.connect('task1_7.db')  
c = conn.cursor()

request = """DROP TABLE IF EXISTS table1;"""
c.execute(request)

request = """CREATE TABLE IF NOT EXISTS table1(column1 int, column2 int, column3 int)"""
c.execute(request)

column1, column2, column3= [], [], []
for i in range(1000):
    column1.append(randint(0, 1000))
    column2.append(randint(0, 1000))
    column3.append(randint(0, 1000))
    
data = {
    "column1": column1,
    "column2": column2,
    "column3": column3
}

df = pd.DataFrame(data)

df.to_sql('table1', con = conn, if_exists = 'append', index = False)
request = """SELECT * FROM table1"""
df = pd.read_sql_query(request, conn)
conn.close()

df
```

```python
#Далее построить гистаграмму распределения этих трех колонок

fig = px.histogram(df)
fig.show()
```
